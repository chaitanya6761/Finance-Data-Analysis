# Finance Data Analysis 

In this data project we will focus on some exploratory data analysis and visualisations of stock prices. 

We'll focus on bank stocks and see how they progressed throughout the [financial crisis](https://en.wikipedia.org/wiki/Financial_crisis_of_2007%E2%80%9308) all the way to early 2016.

## Data And Setup

We will use the stock information of the following banks for analysis:
*  Bank of America
* CitiGroup
* Goldman Sachs
* JPMorgan Chase
* Morgan Stanley
* Wells Fargo


```python
import pandas as pd
import numpy as np
import seaborn as sns
sns.set_style('whitegrid')
import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')

bank_data = pd.read_pickle('all_banks')
print('The Size Of The DataFrame: ',bank_data.shape)
```

```python
bank_data.head()
```

```python
#List Of Bank Ticker Symbols
tickers = list(bank_data.columns.levels[0])
tickers
```

# EDA

Let's explore the data a bit! Before continuing, We will be using [Using .xs](http://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.xs.html) to traverse our dataframe.


**What is the max volumes traded for ecah bank's share throughout the time period?**

```python
bank_data.xs(key='Volume',level='Stock Info', axis=1).max()
```

**What are the Highest and Lowest prices hit by each stock thoughout the time period ?**

```python
#Highest Prices
data = bank_data.xs(key='High',level='Stock Info', axis=1).max()
index = bank_data.xs(key='High',level='Stock Info', axis=1).idxmax()
pd.DataFrame({'Date':index,'Price':data})
```

```python
#Lowest Prices
data = bank_data.xs(key='Low',level='Stock Info', axis=1).min()
index = bank_data.xs(key='Low',level='Stock Info', axis=1).idxmin()
pd.DataFrame({'Date':index,'Price':data})
```

- **The above two dataframes show that the lowest prices were hit by the stocks were during 2008-09 financial crisis.**

**What is the max Close price for each bank's stock throughout the time period?**

```python
data = bank_data.xs(key='Close',level='Stock Info', axis=1).max()
index = bank_data.xs(key='Close',level='Stock Info', axis=1).idxmax()
pd.DataFrame({'Date':index,'Price':data})
```

**Lets Create a new empty DataFrame called returns. This dataframe will contain the returns for each bank's stock. returns are typically defined by:**

$$r_t = \frac{p_t - p_{t-1}}{p_{t-1}} = \frac{p_t}{p_{t-1}} - 1$$

**We can use pandas pct_change() method on the Close column to create a column representing this return value. Create a for loop that goes and for each Bank Stock Ticker creates this returns column and set's it as a column in the returns DataFrame.**

```python
returns = pd.DataFrame()
for ticker in tickers: 
    column_name = ticker + ' Return'
    returns[column_name] = bank_data[ticker]['Close'].pct_change()
    
returns.head()
```

**Let's create a pairplot using seaborn of the returns dataframe. What stock stands out !!**

```python
sns.pairplot(returns[1:])
```

- **All other stocks seem to be co related, but citi group stock stands out with wierd line of dots when compared with other stocks, because it's stock suffered a huge crash.**

**Using this returns DataFrame, Lets figure out on what dates each bank stock had the best and worst single day returns.**

```python
#Worst Returns
data = returns.min()
index = returns.idxmin()
pd.DataFrame({'Date':index,'Return':data})
```

- **Four out of six stocks have same day as worst drop. Acutally that day was Barack Obama's inaugural address, So some big investors might got panicky and withdrew their stocks from market, which led to big fall in stock prices.**

```python
#Best Returns
data = returns.max()
index = returns.idxmax()
pd.DataFrame({'Date':index,'Return':data})
```

- **Citi Group's largest drop and biggest gain are very close to each other because during those dates, citi group annouced their reverse stock split.**

**Lets take a look at the standard deviation of the returns, which stock would we classify as the riskiest over the entire time period? Which would we classify as the riskiest for the year 2015?**

```python
returns.std()
```

```python
returns.loc['2015-01-01':].std()
```

- **As expected citi group is the most riskiest because it has a hight std.** 
- **All stocks seem to have almost the same risk profile for the year 2015, with MS having the highest std.**

**Lets create a distplot using seaborn of the 2008 returns for CitiGroup**

```python
plt.figure(figsize=(10,6))
sns.distplot(returns.loc['2008-01-01':'2008-12-31']['C Return'], kde=False, color='red')
```

**Lets create a distplot using seaborn of the 2015 returns for Morgan Stanley**

```python
plt.figure(figsize=(10,6))
sns.distplot(returns.loc['2015-01-01':'2015-12-31']['MS Return'], color='green', bins=50, kde=False)
```
